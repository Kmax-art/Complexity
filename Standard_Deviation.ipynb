```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import block_diag
from numpy import linalg as LA
import pandas as pd
from scipy import integrate
from scipy.linalg import logm, expm
from time import time
from Complexity_Functions import *
# import seaborn as sns
```

```python
sample_size = 1000
Ψ_CP = DQW(np.pi/4)
step = 10
U12 = SAMPLE_GEN(Ψ_CP[step],Ψ_CP[step+1],sample_size)
niel_com = np.zeros(sample_size,dtype = float)

for i in range(sample_size):    
    niel_com[i] = np.real(COMPLEXITY(U12[i],2))
```

```python
deviation = np.zeros_like(niel_com,dtype = 'float')

for i in range(len(deviation)):
    deviation[i] = np.std(np.real(niel_com[0:i+1]))
```

```python
xx = np.linspace(1,sample_size,sample_size)
```
